```python
import math
def funcion1(cant_Elementos):

    suma=0
    lista=[]
    for i in range(cant_Elementos+1):
        
        suma= suma + (1/(2*i+1))*pow(-1,i)
        lista.append(suma)

    return lista,suma
```

```python
def funcion2(referencia,precision):
    
    centinela=1
    suma=0
    i=0
    lista=[]
    while(centinela):
        
        suma= suma + (1/(2*i+1))*pow(-1,i)
        lista.append(suma)
        
        if(abs(suma - referencia)< precision):

            centinela=0
              
        i += 1
    return lista,suma
```

```python
def funcion3(precision):
    
    centinela=1
    suma=0
    i=0
    lista=[]
    while(centinela):
        suma_anterior= suma
        suma= suma + (1/(2*i+1))*pow(-1,i)
        lista.append(suma)
        
        if(abs(suma-suma_anterior)<=precision):

            centinela=0
              
        i += 1
    return lista,suma
```
