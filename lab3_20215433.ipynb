```python
%%file  main.c

#include <time.h>
#include <stdio.h>
#include <stdlib.h>
#include <x86intrin.h>
#include <math.h>

//3.1 Funcion que calcula la mediana de un arreglo de elementos tipo double:

double calcular_mediana(long* arr, int N){

    //Primero debemos ordenar los valores del arreglo (De menor a mayor)

    double temp;
    for(int i = 0; i < N - 1; i++){
        for(int j = 0; j < N - i - 1; j++){
            if (arr[j] > arr[j+1]){
                temp = arr[j];
                arr[j] = arr[j+1];
                arr[j+1] = temp;
            }
        }
    }

    if (N % 2 == 0){
        return ((double)arr[ N / 2 ] + (double)arr[ (N/2) - 1 ]) / 2.0;
    }
    else {
        return arr[ N / 2 ];
    }
    
}


//3.2 Funciones del item 2

int suma(int a, int b){
    return a+b;
}

int resta(int a, int b){
    return a-b;
}

int producto(int a, int b){
    return a*b;
}

int suma_cuadrados(int a, int b){
    return a*a+b*b;
}

int suma_cubos(int a, int b){
    return a*a*a+b*b*b;
}

int suma_cuartas(int a, int b){
    return a*a*a*a+b*b*b*b;
}

//3.3 Función que devuelve el número de instrucciones de las funciones del item 2

int numero_instrucciones(int n){

    switch(n){

        case 1:
            return 3;
        
        case 2:
            return 4;

        case 3:
            return 4;
            
        case 4:
            return 5;

        case 5:
            return 9;

        case 6:
            return 7;
    }
}

//3.4 Funcion que devuelve el CPI

double calcular_cpi(double cant_ciclos, int cant_instrucciones){

    return cant_ciclos/cant_instrucciones;

}

//3.5 Función que calcula los MIPS

double calcular_mips(double frecuencia, double cpi) {
    return (double)frecuencia/(cpi*pow(10,6));


}

//3.6 Funcion CPI_mediano

double CPI_mediano(int numero_entero){

    srand(time(NULL));

    int a;
    int b;
    int res;

    long tic;
    long toc;
    
    
    int N = 250000;


    //Usamos long porque la diferencia entre ciclos suelen ser numeros podría ser muy grande, para evitar imprevistos usamos long.

    long* arr = (long*)calloc(N, sizeof(long));
    for(int i = 0; i < N; i++){

        //Usaremos números random del 0 al 8

        a = rand()%9;
        b = rand()%9;

        // Hacemos una selectiva para saber para qué operación se va a trabajar dependiendo el "numero entero"
        if(numero_entero==1){
            tic = __rdtsc();
            res = suma(a,b);
            toc = __rdtsc();
        
        }
        else if(numero_entero==2){
            tic = __rdtsc();
            res = resta(a,b);
            toc = __rdtsc();
        }
        else if(numero_entero==3){
            tic = __rdtsc();
            res = producto(a,b);
            toc = __rdtsc();
        }
        else if(numero_entero==3){
            tic = __rdtsc();
            res = suma_cuadrados(a,b);
            toc = __rdtsc();
        }
        else if(numero_entero==3){
            tic = __rdtsc();
            res = suma_cubos(a,b);
            toc = __rdtsc();
        }
        else{
            tic = __rdtsc();
            res = suma_cuartas(a,b);
            toc = __rdtsc();
        }
        
        // guardamos las mediciones en un vector arr[]
        arr[i] = (long)(toc - tic);
    }

    //obtenemos el números de instrucciones de la función 3.3
    int num_intrucciones= numero_instrucciones(numero_entero);
    
    //calculamos la mediana con la función 3.1
    double mediana= calcular_mediana(arr,N);

    //finalmente con la función 3.4 calculamos el CPI mediano, este tiene de argumentos la mediana de todas las mediciones 
    //y el numero de intrucciones dependiendo la operación

    double cpi_mediano = calcular_cpi(mediana,num_intrucciones);


    free(arr);
    return cpi_mediano;


}

//3.7 ¿Qué media pitagórica debe emplear para encontrar el CPI medio de los CPIs calculados?
//Usamos la media aritmética pues los valores de cpi no van a estar muy alejados y estos son independientes


//3.9 Función que calcula la media aritmética generalizada, este recibe como argumentos un arreglo en donde están almacenados el CPI de cada operación
double calcular_media_cpi(double* arr_media,int N){

    double suma=0.0;
    
    for (int i=0;i<N;i++){
        
        suma=suma+arr_media[i];

    }
    return suma/N;

}


//3.9 Función main
int main(){

    //Uso la frecuencia de mi computador, 2300 MHz
    double frecuencia =2300*pow(10,6);

    int N=6;

    //Usamos una iterativa for para imprimir los CPI mediano de cada operación y además se usará un vector para tener almacenados los CPI de cada función

    double* arr_media = (double*)calloc(N, sizeof(double));

    for(int i=1;i<=6;i++){

        arr_media[i-1]=CPI_mediano(i);

        printf("CPI mediano calculado para la funcion %d es %lf \n",i,arr_media[i-1]);


    }
    
    //Calculamos la media aritmética del CPI de cada operación ( en total 6)

    double media_cpi = calcular_media_cpi(arr_media,N);

    //Calculamos indicador MIPS
    double mips = calcular_mips(frecuencia,media_cpi);

    //Impresiones de la media y MIPS

    printf ("Media pitagóricas de las mediciones(media aritmética): %lf \n",media_cpi);
    printf ("Mips : %lf",mips);

    free(arr_media);

    return 0;
}







```
